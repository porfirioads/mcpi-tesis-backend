# An√°lisis de sentimientos con dataset personalizado

## Lectura del dataset

```python
import pandas as pd

df = pd.read_csv(r'sample-porfirio-separado.csv', encoding='latin-1')
```

```python
df2 = df.copy()
df2 = df2[['Answer', 'Sentiment']]
df2.drop(df2[df2['Sentiment'] == 'Neutral'].index, inplace=True)
df2['Sentiment'] = df2['Sentiment'].replace({'Positive': 'pos', 'Neutral': 'neu', 'Negative': 'neg'})
df2
```

## Creaci√≥n de datos de entrenamiento y de prueba

```python
from sklearn.model_selection import train_test_split

y = df2['Sentiment']
X_train, X_test, y_train, y_test = train_test_split(df2, y, test_size=0.25, random_state=25)
```

## An√°lisis de datos

```python
from textblob.classifiers import NaiveBayesClassifier
cl = NaiveBayesClassifier(X_train.values)
cl.accuracy(X_test.values)
```

```python
from textblob.classifiers import DecisionTreeClassifier
cl2 = DecisionTreeClassifier(X_train.values)
cl2.accuracy(X_test.values)
```

```python
from textblob.classifiers import MaxEntClassifier
cl3 = MaxEntClassifier(X_train.values)
cl3.accuracy(X_test.values)
```

```python
from textblob.classifiers import NLTKClassifier
cl4 = NLTKClassifier(X_train.values)
cl4.accuracy(X_test.values)
```

```python
from textblob.classifiers import PositiveNaiveBayesClassifier
cl5 = PositiveNaiveBayesClassifier(X_train.values)
cl5.accuracy(X_test.values)
```
